```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file
data = pd.read_csv("CO2 vs GDP.csv")


# Clean the GDP column: remove commas and convert to numeric
data['India GDP (Bill USD)'] = pd.to_numeric(data['India GDP (Bill USD)'].replace(',', '', regex=True), errors='coerce')

# Drop rows with NaN values in either column
cleaned_data = data.dropna(subset=['Annual CO2 emissions', 'India GDP (Bill USD)'])

# Extract cleaned columns
co2_emissions_cleaned = cleaned_data['Annual CO2 emissions']
gdp_cleaned = cleaned_data['India GDP (Bill USD)']
# Calculate the trend line
z = np.polyfit(gdp_cleaned, co2_emissions_cleaned, 1)  # Linear regression (degree=1)
p = np.poly1d(z)

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(gdp_cleaned, co2_emissions_cleaned, color='blue', alpha=0.7, label='Data Points')

# Add the trend line
plt.plot(gdp_cleaned, p(gdp_cleaned), color='red', linewidth=2, label='Trend Line')

# Add titles and labels
plt.title('Scatter Plot with Trend Line: CO2 Emissions vs GDP', fontsize=14)
plt.xlabel('GDP (Billions USD)', fontsize=12)
plt.ylabel('Annual CO2 Emissions', fontsize=12)

# Add a legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
